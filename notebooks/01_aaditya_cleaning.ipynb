```python
#check for the outliners
df.sort_values(by='pages',inplace=True)
def checkcoutlinears(df,col):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower =Q1-1.5*IQR
    upper=Q3+1.5*IQR
    outliner=df[(df[col]<lower) | (df[col]>upper)]
    return outliner
```

```python
outliners=checkcoutlinears(df,'pages')
```
