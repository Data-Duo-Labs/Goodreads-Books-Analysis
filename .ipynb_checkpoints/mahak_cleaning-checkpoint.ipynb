```python
df['formatCategory'] = df['bookFormat']
filt_other_categories = ~ (df['formatCategory'].str.contains("Hardcover|Paperback|Kindle|eBook|Audiobook"))
df.loc[filt_other_categories, 'formatCategory'] = "Others"
```
